```python
def make_bins(idx_, bins_):
    d_=round(len(idx_)/bins_)
    splits=[idx_[i*d_:(i+1)*d_] for i in range(bins_-1)]
    splits.append(idx_[(bins_-1)*d_:])
    return splits
```

```python
def shuffle_bins(xx):
    shuff_xx=[]
    for x in xx:
        sub_shuff=random.sample(x, len(x))
        shuff_xx.extend(sub_shuff)
    return shuff_xx
```

```python
def minimal_row_coverage_sets(XX, splits):
    idx_shuffle=shuffle_bins(splits)

    z=XX[:,idx_shuffle[0]]
    idx_values=[idx_shuffle[0]]
    coverages=[]
    i=1

    while i<len(idx_shuffle): 
        d0=z.nnz #l0 norm of z=number of nonzeros=row coverage
        
        if d0==m_: #achieved full coverage of rows!
            print('whippee, early stop at iter '+str(i))
            break
        else: # compare next two vectors and see which increases row coverage
            d1=(z+XX[:,idx_shuffle[i]]).nnz #l0 norm of sum vector plus next x1
            d2=(z+XX[:,idx_shuffle[i+1]]).nnz #l0 norm of sum vector plus next x2

            if d0 >= d1 and d0 >=  d2: # the next two vectors do not expand row coverage
                coverages.append(d0)

            elif d1 >= d2: # if x1 increases range more than (or equivalent) to x2, then keep x1 and make the sum z 
                z=z+XX[:,idx_shuffle[i]]
                idx_values.append(idx_shuffle[i])
                coverages.append(d1)

            else: # vice versa
                z=z+XX[:,idx_shuffle[i+1]]
                idx_values.append(idx_shuffle[i+1])
                coverages.append(d2)
            
            i+=2 # jump by 2 each time

    return len(idx_values), idx_values, coverages
```
